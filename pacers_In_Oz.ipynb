```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data (update the file path if needed)
file_path = 'Bowlers_In_OZ.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')

# Ensure numerical columns are properly parsed
data['Bowl Ave'] = pd.to_numeric(data['Bowl Ave'], errors='coerce')
data['SR'] = pd.to_numeric(data['SR'], errors='coerce')

# Drop rows with missing values in key columns
data = data.dropna(subset=['Bowl Ave', 'SR'])

# Calculate mean values for equal quadrants
x_mean = data['Bowl Ave'].mean()
y_mean = data['SR'].mean()

# Determine the range for both axes
x_range = data['Bowl Ave'].max() - data['Bowl Ave'].min()
y_range = data['SR'].max() - data['SR'].min()

# Find the larger range to ensure the quadrants are equal
max_range = max(x_range, y_range)

# Set axis limits based on the larger range
x_limit = [x_mean - max_range / 2, x_mean + max_range / 2]
y_limit = [y_mean - max_range / 2, y_mean + max_range / 2]

# Plot the quadrant chart
plt.figure(figsize=(10, 7))

# Scatter plot for players
plt.scatter(data['Bowl Ave'], data['SR'], color='blue', s=50)

# Add vertical and horizontal lines for quadrants
plt.axvline(x=x_mean, color='red', linestyle='--', label=f'X Mean: {x_mean:.2f}')
plt.axhline(y=y_mean, color='green', linestyle='--', label=f'Y Mean: {y_mean:.2f}')

# Annotate player names
for i, player in enumerate(data['Player']):
    plt.text(data['Bowl Ave'][i] + 0.1, data['SR'][i] + 0.1, player, fontsize=8)

# Set equal axis limits to ensure equal quadrants
plt.xlim(x_limit)
plt.ylim(y_limit)

# Chart labels and title
plt.title('20th and 21st Century - Pacers in Australia - Min 37 wkts', fontsize=16)
plt.xlabel('Bowling Average (Bowl Ave)', fontsize=12)
plt.ylabel('Strike Rate (SR)', fontsize=12)
plt.legend()

# Show grid and plot
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
summary = data[['Bowl Ave', 'SR']].describe()
print(summary)
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(data[['Bowl Ave', 'SR']]))
outliers = (z_scores > 3).all(axis=1)
print(data[outliers])
```
